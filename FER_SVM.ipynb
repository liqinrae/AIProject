## IMPORT DATA ##

```python
import cv2
import dlib
import numpy as np
import os

print(os.listdir("data/CK+48"))
```

### GET FACE LANDMARK

```python
data_path = 'data/CK+48'
data_dir_list = os.listdir(data_path)
output_path = 'output/cv2-img/'
emotions =['happy', 'contempt', 'fear', 'surprise', 'sadness', 'anger', 'disgust']

img_rows=256
img_cols=256
num_channel=1

num_epoch=10

img_data_list=[]
label_list=[]

#人脸分类器
detector = dlib.get_frontal_face_detector()
# 获取人脸检测器
predictor = dlib.shape_predictor(
    "shape_predictor_68_face_landmarks.dat"
)


for dataset in data_dir_list:
    img_list=os.listdir(data_path+'/'+ dataset)
    emotion_folder=os.mkdir(output_path+dataset)
    print ('Loaded the images of dataset-'+'{}\n'.format(dataset))
    label = emotions.index(dataset)
    i = 0
    for img in img_list:
        feature_pt_list = []
        i = i + 1 
        input_img=cv2.imread(data_path + '/'+ dataset + '/'+ img )
        
        input_img_resize=cv2.resize(input_img,(48,48))
        
        gray=cv2.cvtColor(input_img_resize, cv2.COLOR_BGR2GRAY)
        
        dets = detector(gray, 1)
        for face in dets:
            shape = predictor(input_img_resize, face)  # 寻找人脸的68个标定点
            # 遍历所有点，打印出其坐标，并圈出来
            for pt in shape.parts():
                pt_pos = (pt.x, pt.y)
                cv2.circle(input_img_resize, pt_pos, 2, (0, 255, 0), 1)
            save_path = output_path+dataset+"/"+str(i)+".png"
            #if i == 2:
            
            cv2.imwrite(save_path, input_img_resize)
            print(save_path)
            #pt_mat = np.matrix([[p.x, p.y] for p in shape.parts()])
            for p in shape.parts():
                feature_pt_list.append(p.x)
                feature_pt_list.append(p.y)
                

        img_data_list.append(feature_pt_list)
        label_list.append(label)
        
        
print("THE END")
```

### SPLIT DATA TO TRAINING SET & TEST SET

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(img_data_list, label_list, test_size=0.3,random_state=109) 
```

### TRAIN THE MODEL 

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') 
clf.decision_function_shape = "ovr"

#Train the model using the training sets
clf.fit(X_train, y_train)
```
